#### Introduction to the OpenAI Chat Completions API

```python
import os
from openai import OpenAI

api_key =  ''
base_url = "https://api.openai.com/v1"

# Create an OpenAI client with an API key and base URL
client = OpenAI(
    api_key=api_key,
    base_url=base_url,
)
```

```python
response = client.chat.completions.create(
    model="gpt-3.5-turbo",    
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Help me write a hook for a new blog post about our Ethiopian dark roast."},
  ]
)
print(response)
```

```python
response = client.chat.completions.create(
    model="gpt-3.5-turbo",    
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Help me write a hook for a new blog post about our Ethiopian dark roast."},
  ]
)
api_call_id = response.id 
model_used = response.model
total_tokens = response.usage.total_tokens
ai_output = response.choices[0].message.content

print(api_call_id)
print(model_used)
print(total_tokens)
print(ai_output)
```

```python
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    response_format={ "type": "json_object" },
    messages=[
        {"role": "system", "content": "You are a marketing assistant for GlobalJava Roasters crafting marketing materials in JSON format."},
        {"role": "user", "content": "Generate a blog post preview about the benefits of our new Guatamalan coffee blend."}
  ],
  max_tokens=1024  # Safe upper limit 
)   
print(response.choices[0].message.content)
```
